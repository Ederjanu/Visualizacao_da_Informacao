```python
amox_por_Estado = amox.groupby('UF_VENDA', sort=False).agg({'QTD_VENDIDA': np.sum}).reset_index()
amox_por_Estado
```

```python
import plotly.express as px
```

```python
px.bar(amox_por_Estado, x = 'UF_VENDA', y = 'QTD_VENDIDA')
```

```python
from numpy.lib.polynomial import RankWarning
px.bar(amox_por_Estado, y = 'UF_VENDA', x = 'QTD_VENDIDA', orientation='h', color ='UF_VENDA')
```

```python
Quantidade_de_venda_por_Estado = amox['UF_VENDA'].value_counts()
Quantidade_de_venda_por_Estado
```

```python
px.pie(Quantidade_de_venda_por_Estado, names = Quantidade_de_venda_por_Estado.index, values = Quantidade_de_venda_por_Estado.values)
```
